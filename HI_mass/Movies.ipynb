This notebook shows how to make a small movie from the subcubes.
`ffmpeg` need to be downloaded first.

```python
from spectral_cube import SpectralCube
import matplotlib
from matplotlib import rcParams
import matplotlib.animation as animation
import pylab as pl
rcParams['animation.writer'] = 'ffmpeg'
rcParams['animation.ffmpeg_path'] = '/usr/local/bin/ffmpeg'
print(matplotlib.animation.writers.list())
```

```python
ts6 = SpectralCube.read('Subcube7.fits')
n6 = SpectralCube.read('Noisecube7.fits')
```

### Subcube

```python
'''
Animation Setup Cell. Change with Care.
'''
pl.figure(figsize=(5,5))
fig, ax = pl.subplots()

im = ax.imshow(ts6[0,::1,:].array,vmin=-0.1e-3, vmax=1e-3,cmap='magma')
ax.set_xticks(())
ax.set_yticks(())
ax.invert_yaxis()
num_frames = len(ts6)

# How long each frame be in milli-seconds?
interval=200.0

def init():
    im.set_data(ts6[0,::1,:].array)
    return (im,)

def animate(i):
    ts_slice = ts6[i,::1,:].array
    im.set_data(ts_slice)
    return(im,)

anim = animation.FuncAnimation(fig, animate, init_func=init,
                              frames=num_frames, interval=interval, 
                               blit=True)
```

```python
from IPython.display import HTML
HTML(anim.to_html5_video()) #then we can just convert the video to GIF
```
